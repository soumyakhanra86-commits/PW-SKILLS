Q.5/ How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?  
ANS:- The Cumulative Distribution Function (CDF), $F(x)$, gives the total accumulated probability up to a certain value x.$F(x) = P(X \le x)$It works for both discrete and continuous variables.For a continuous variable, the CDF is the integral (total area) of the PDF from $-\infty$ to x. The PDF is the derivative (rate of change) of the CDF

Q.15/ What is the application of Z statistics in hypothesis testing ?  
ANS:- Z-statistics are used in hypothesis testing when:
Population standard deviation is known.
Sample size is large (n ≥ 30).
They help decide whether to accept or reject a null hypothesis.

Q.16/ How do you calculate a Z-score, and what does it represent ?  
ANS:- A Z-score is calculated by taking a raw data point, subtracting the population mean, and then dividing the result by the population standard deviation. It represents the number of standard deviations a given data point is away from the mean .

Q.17/ What are point estimates and interval estimates in statistics ?  
ANS:- Point estimate:- When a singel number is used to estimate an unknown population it is called point estimate.    

interval estimates:-A range of values that is likely to contain the unknown population parameter. This is far more informative as it includes a measure of uncertainty.

Q.18/What is the significance of confidence intervals in statistical analysis ?   
 ANS:- The significance of cofidence intervals in statistical analysis is that they provide a possible range of values for an unknown population parameter, along with a quantified measure of the uncertainty associated with the estimate .

Q.19/ What is the relationship between a Z-score and a confidence interval ?  
ANS:-The Z-score and the confidence interval (CI) are closely related because Z-scores are used as critical values to calculate the margin of error, which defines the width of the confidence interval .

 Q.20/ How are Z-scores used to compare different distributions ?  
 ANS:- Z-scores are primarily used to compare observations from different distributions by standardizing them onto a common, universal scale: the Standard Normal Distribution.This standardization process allows for a direct, apples-to-apples comparison of relative standing, even if the original distributions have vastly different means and standard deviations

Q.22/ What is the concept of expected value in a probability distribution ?  
ANS:- The expected value of a random variable is the theoretical mean or long-run average of all possible outcomes in a probability distribution.It tells us what we’d expect to happen on average if an experiment is repeated many times.

Q.23/ How does a probability distribution relate to the expected outcome of a random variable ?
