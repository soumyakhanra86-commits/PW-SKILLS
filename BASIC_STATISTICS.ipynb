Q.1/How do you calculate the mean, median, and mode of a dataset ?

```python
import numpy as np
from scipy import stats

data = [45, 50, 65, 70, 70, 80, 90]

mean_value = np.mean(data)
median_value = np.median(data)
mode_result = stats.mode(data, keepdims=True)
mode_value = mode_result.mode[0]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

Q.4/ Implement sampling techniques like random sampling and stratified sampling ?

```python
import pandas as pd
data = {
    'Department': ['HR', 'HR', 'HR', 'HR', 'HR',
                   'Sales', 'Sales', 'Sales', 'Sales', 'Sales', 'Sales', 'Sales',
                   'IT', 'IT', 'IT', 'IT', 'IT', 'IT', 'IT', 'IT'],
    'Salary': [50, 55, 60, 52, 58, 70, 75, 80, 72, 77, 85, 78, 90, 95, 100, 88, 92, 110, 105, 98]
}
df = pd.DataFrame(data)

print("--- Original DataFrame ---")
print(f"Total size: {len(df)}")
print(df['Department'].value_counts())
print("\n")

simple_random_sample = df.sample(n=6, random_state=42)
print("--- Simple Random Sample (n=6) ---")
print(simple_random_sample)
print("\n")

stratified_sample = df.groupby('Department', group_keys=False).sample(n=2, random_state=42)
print("--- Stratified Sample (n=2 from each group) ---")
print(stratified_sample)
```

Q.5/Write a Python function to calculate the range of a dataset ?

```python
def calculate_range(data):
    minimum = min(data)
    maximum = max(data)
    return maximum - minimum

my_data = [10, 12, 15, 11, 13, 30, 5]
data_range = calculate_range(my_data)
print(f"Dataset: {my_data}")
print(f"Maximum: {max(my_data)}")
print(f"Minimum: {min(my_data)}")
print(f"Range: {data_range}")
```

Q.13/Calculate the mean, median, and mode of grouped data ?

```python
classes = [0,10,20,30,40,50]
freq = [5,9,12,8,6]
midpoints = [(classes[i]+classes[i+1])/2 for i in range(len(freq))]
mean_grouped = sum(midpoints[i]*freq[i] for i in range(len(freq))) / sum(freq)
print("Grouped Mean:", mean_grouped)
```

Q.14/ Simulate data using Python and calculate its central tendency and dispersion ?

```python
import random
from scipy import stats
import numpy as np

data = [random.randint(10, 100) for _ in range(20)]
mean = sum(data)/len(data)
sorted_data = sorted(data)
if len(data)%2==0:
    median = (sorted_data[len(data)//2 - 1] + sorted_data[len(data)//2]) / 2
else:
    median = sorted_data[len(data)//2]

mode_result = stats.mode(data)
if hasattr(mode_result, 'mode') and np.ndim(mode_result.mode) > 0:
    mode = mode_result.mode[0]
else:
    mode = mode_result.mode

variance = sum((x - mean) ** 2 for x in data) / len(data)
std_dev = variance ** 0.5
print("Data:", data)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Range:", max(data)-min(data))
```

Q.19/ Compare two datasets using their standard deviations ?

```python
import pandas as pd
data1 = [45, 50, 55, 60, 65, 70, 75]
data2 = [30, 45, 60, 75, 90, 105, 120]
df = pd.DataFrame({
    'Dataset_1': data1,
    'Dataset_2': data2
})
std1 = df['Dataset_1'].std()
std2 = df['Dataset_2'].std()
print("Standard Deviation of Dataset 1:", std1)
print("Standard Deviation of Dataset 2:", std2)
if std1 > std2:
    print("Dataset 1 is more spread out.")
elif std2 > std1:
    print("Dataset 2 is more spread out.")
else:
    print("Both datasets have equal spread.")
```

Q.20/ Write a Python program to visualize covariance using a heatmap ?

Q.23/Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn ?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
np.random.seed(0)
data = np.random.exponential(scale=2, size=1000)
skewness = skew(data)
kurt = kurtosis(data)
print("Skewness:", round(skewness, 2))
print("Kurtosis:", round(kurt, 2))
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.histplot(data, kde=True, color="skyblue", bins=30)
plt.title(f"Distribution (Skewness = {skewness:.2f}, Kurtosis = {kurt:.2f})")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Q.24/ Implement the Pearson and Spearman correlation coefficients for a dataset?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
data = {
    'X': np.arange(1, 11),
    'Y': np.arange(1, 11) + np.random.normal(0, 1, 10),
    'Z': np.random.randint(1, 20, 10)
}
df = pd.DataFrame(data)
print("Dataset:\n", df)
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')
print("\n📈 Pearson Correlation Coefficient:")
print(pearson_corr)
print("\n📊 Spearman Correlation Coefficient:")
print(spearman_corr)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.tight_layout()
plt.show()
```
