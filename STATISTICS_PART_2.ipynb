Q.1/ What is hypothesis testing in statistics ?  
ANS:- Hypothesis testing is a statistical method used to decide wether there is enough evidence in a sample to support a claim about a population.

Q.2/ What is the null hypothesis, and how does it differ from the alternative hypothesis ?  
ANS:- The null hypothesis states that there is no effect or no difference.
The alternative hypothesis states that an effect or difference does exist .

Q.3/ What is the significance level in hypothesis testing, and why is it important?   
ANS:-The significance level is the probability of rejecting a true null hypothesis . alpha=0.05.  
It is important because it sets the decision thrershold for accepting or rejecting a null hypothesis.

Q.4/ What does a P-value represent in hypothesis testing ?  
ANS:-p-value represents the probability of observing the sample results, or something more extreme, assuming the null hypothesis is true. It measures the strength of evidence against null hypothesis.

Q.5/ How do you interpret the P-value in hypothesis testing ?  
ANS:- If p ≤ alpha, there is strong evidence against null hypothesis, so we reject it.
If p > alpha, we fail to reject null hypothesis because the evidence is weak .

Q.6/ What are Type 1 and Type 2 errors in hypothesis testing ?  
ANS:- A type 1 error happens when a true null hypothesis is wrongly rejected .  
A type 2 error happenas  when a a false null hypothesis is not rejected.

Q.7/ What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?  
ANS:-A one-tailed test checks for an effect in a specific direction.
A two-tailed test checks for any difference, regardless of direction.
Two-tailed tests are more conservative.

Q.8/ What is the Z-test, and when is it used in hypothesis testing ?  
ANS:- A z-test is used to test hypothesis about population means when the population standerd deviation is known.
It is typically applied to large samples (n ≥ 30).

Q.9/ How do you calculate the Z-score, and what does it represent ?  
ANS:- Z-score = (sample value - mean) / standard deviation.
It represents how many standard deviations a value is from the mean.
Higher absolute Z-scores show more extreme values .

Q.10/ What is the T-distribution, and when should it be used instead of the normal distribution ?  
ANS:-The t-distribution is similar to the normal distribution but has heavier tails.
It is used when the sample size is small (n < 30) and population standard deviation is unknown.
It provides a more accurate estimation for small samples

Q.11/ What is the difference between a Z-test and a T-test ?  
ANS:- A ztest is used when sample size is large and standerd deviation is known.
T test is used when sample size is small and standard deviation is unkown.

Q.12/ What is the T-test, and how is it used in hypothesis testing ?  
ANS:- A T-test compares sample means to determine whether differences are statistically significant.
It is used for small samples with unknown population SD.
Common types include one-sample, independent, and paired T-tests.

Q.13/ What is the relationship between Z-test and T-test ?  
ANS:-T-test and Z-test are both test hypothesis about population means .  
The T-test becomes very similar to the Z-test when the sample size becomes large.They differ mainly in whether population standard deviation is known.


Q.14/ What is a confidence interval, and how is it used ?  
ANS:- A confidence interval provides a range of values that likely contains the true population parameter.It helps measure the precision of an estimate.

Q.15/ What is the margin of error, and how does it affect the confidence interva ?   
ANS:-The margin of error is the range of values added to and subtracted from the sample estimate to create the confidence interval .

Q.16/How is Bayes' Theorem used in statistics, and what is its significance ?  
ANS:-Bays theorem updates the probability of an event based on new information.
It links prior probability with likelihood to produce a posterior probability.

Q.17/ What is the Chi-square distribution, and when is it used ?  
ANS:-The chi-square distribution is used for categorical data analytics. it test relationship between variabels or whether observed counts match expected counts.

Q.18/ What is the Chi-square goodness of fit test, and how is it applied ?  
ANS:-It checks if observed data matches an expected theoritica distribution.
It compares observed and expected frequencies using the Chi-square formula .

Q.19/ What is the F-distribution, and when is it used in hypothesis testing ?  

ANS:- The F-distribution is a continuous probability distribution that is also non-symmetric and only takes positive values. It is defined by two degrees of freedom: the numenetor and denominetor.

Q.20/ What is an ANOVA test, and what are its assumptions?  
ANS:-ANOVA is a statistical technique used to compare the means of three or more independent groups to determine if at least one group mean is statistically different from the others.  
Assumptions include normality, equal variances, and independent observations.
It checks whether group differences are statistically significant.

Q.21/ What are the different types of ANOVA tests ?

Q.11/ Simulate multiple samples from a normal distribution and verify the Central Limit Theorem ?

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def verify_clt(mu=5, sigma=2, repeats=20000, sample_sizes=[1,5,10,30,100]):
    for n in sample_sizes:
        samples = np.random.normal(mu, sigma, size=(repeats, n))
        sample_means = samples.mean(axis=1)

        theoretical_std = sigma / math.sqrt(n)

        plt.hist(sample_means, bins=40, density=True, alpha=0.6)
        x = np.linspace(sample_means.min(), sample_means.max(), 300)
        pdf = (1/(theoretical_std*np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mu)/theoretical_std)**2)
        plt.plot(x, pdf)
        plt.title(f"CLT Verification (n={n})")
        plt.show()

verify_clt()
```

Q.1/ Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results ?

```python
import numpy as np
from scipy.stats import norm
sample = np.array([52, 55, 50, 49, 51, 53, 54, 52, 50, 51])
pop_mean = 50
pop_std = 5
sample_mean = np.mean(sample)
n = len(sample)
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print("Z-score:", z)
print("P-value:", p_value)
if p_value < 0.05:
    print("Reject H0: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject H0: No significant difference.")
```

Q.5/Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def visualize_errors(mu0=50, mu1=55, sigma=10, alpha=0.05):
    x = np.linspace(30, 80, 500)
    y0 = norm.pdf(x, mu0, sigma)
    y1 = norm.pdf(x, mu1, sigma)
    zcrit = norm.ppf(1 - alpha)
    cutoff = mu0 + zcrit * sigma
    plt.plot(x, y0, label="H0 True")
    plt.plot(x, y1, label="H1 True")
    plt.axvline(cutoff, color='red')
    plt.fill_between(x, y0, where=(x > cutoff), alpha=0.3, label="Type 1 Error")
    plt.fill_between(x, y1, where=(x < cutoff), alpha=0.3, label="Type 2 Error")
    plt.legend()
    plt.show()
visualize_errors()
```

Q.9/  Write a Python function to calculate the confidence interval for a sample mean and explain its significance ?

```python
import numpy as np
from scipy.stats import norm
def ci_mean(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data, ddof=1)/np.sqrt(len(data))
    z = norm.ppf((1 + confidence)/2)
    return (mean - z*se, mean + z*se)
print(ci_mean([10,12,11,9,13]))
```

Q.10/ Write a Python program to calculate the margin of error for a given confidence level using sample data ?

```python
import numpy as np
from scipy.stats import norm
def margin_of_error(data, confidence=0.95):
    se = np.std(data, ddof=1)/np.sqrt(len(data))
    z = norm.ppf((1 + confidence)/2)
    return z * se
print(margin_of_error([10,11,12,14,13]))
```

Q.26/  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results ?

```python
from scipy.stats import chisquare
import numpy as np

observed = np.random.randint(15, 30, 5)
expected = np.full(5, np.mean(observed))

chi, p = chisquare(observed, expected)
print("Chi2:", chi, "P:", p)
```

Q.27/ Perform a Chi-square test for goodness of fit with simulated data and analyze the results ?

```python
import numpy as np
from scipy.stats import chisquare
np.random.seed(42)
observed = np.random.randint(20, 50, 4)
expected = np.full(4, np.mean(observed))
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print("\nChi-square Statistic:", chi_stat)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject H0 → Observed data does NOT fit the expected distribution.")
else:
    print("\nConclusion: Fail to reject H0 → Observed data fits the expected distribution.")
```
