Q.1/ How do you create a 2D NumPy array and calculate the sum of each row ?

```python
import numpy as np
l1 = np.array([[1, 2],
               [3, 4]])
print("Array:\n", l1)
d = l1.ndim
print("DIMENSION:", d)
row_sum = np.sum(l1, axis=1)
print("Row-wise Sum:", row_sum)
```

Q.4/ A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "Math": [85, 90, 78, 92, 88],
    "Physics": [80, 85, 75, 89, 86],
    "Chemistry": [82, 87, 79, 90, 84],
    "Biology": [78, 83, 76, 88, 82]
}
df = pd.DataFrame(data)
corr_matrix = df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Q.6/ Create a DataFrame and add a new column based on an existing column ?

```python
import pandas as pd
data = {
    "Name": ["Soumya", "Rahul", "Anita"],
    "Marks": [85, 90, 78]
}
df = pd.DataFrame(data)
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 80 else "Fail")
print(df)
```

Q.8/ Create a line plot with multiple lines using Matplotlib ?

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='red', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3', color='green', linestyle='-.', marker='^')
plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

Q.9/ Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold ?

```python
import pandas as pd
data = {
    "Name": ["Soumya", "Amit", "Priya", "Rahul", "Neha"],
    "Age": [18, 25, 30, 16, 22],
    "Marks": [75, 55, 90, 40, 85]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
filtered_df = df[df["Age"] > 20]
print("\nFiltered DataFrame (Age > 20):\n", filtered_df)
```

Q.10/ A Create a histogram using Seaborn to visualize a distribution ?

```python
import seaborn as sns
import matplotlib.pyplot as plt
ages = [18, 20, 22, 21, 19, 25, 30, 32, 31, 28, 26, 24, 22, 20, 19, 18,30]
sns.histplot(ages, bins=8, kde=True, color="skyblue")
plt.title("Age Distribution", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

Q.13/ Create a 3D scatter plot using Plotly ?

```python
import plotly.express as px
import pandas as pd
data = {
    "X": [1, 2, 3, 4, 5],
    "Y": [10, 11, 12, 13, 14],
    "Z": [5, 6, 7, 8, 9],
    "Category": ["A", "B", "A", "B", "A"]
}
df = pd.DataFrame(data)
fig = px.scatter_3d(
    df,
    x="X",
    y="Y",
    z="Z",
    color="Category",
    size="Z",
    symbol="Category"
)
fig.show()
```
